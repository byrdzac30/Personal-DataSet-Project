```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Stephen Curry Stats Oct. 29 2009 to Oct. 29 2018.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.plot(df['Minutes'],df['Score Opponent'],'*', color='green')
plt.xlabel('Minutes Played by Stephen Curry')
plt.ylabel('Opponents Final Score')
plt.title('Impact of Stephen Curry')
```

```python
plt.plot(df['Minutes'],df['Score GS'],'*',color='Blue')
plt.xlabel('Minutes Played by Stephen Curry')
plt.ylabel('Warriors Final Score')
plt.title('Impact of Stephen Curry')
```
